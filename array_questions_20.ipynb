```python
# Given a list of subject names, print each subject with a serial number using a for loop.
subjects = ["maths", "physics", "computer", "social", "science"]

for serial, subject in enumerate(subjects, start=1):
    print(f"{serial}. {subject}")
```

```python
#Find the maximum marks, minimum marks, total, and average from a list of 6 student scores

marks = [45,34,50,29,44,36]

maximum = max(marks)
minimum = min(marks)
total = sum(marks)
average = total/len(marks)

print(f"maximum number in list is : {maximum}")
print(f"minimum number in list is : {minimum}")
print(f"sum of all marks is : {total}")
print(f"average number in list is : {average}")


```

```python
# Given a list of CGPA values, sort them in descending order and find the rank of a given student.
cgpa = [7.7, 8.3, 8.9, 9.8, 9.4, 9.6]

ranked = sorted(cgpa, reverse=True)
print("Ranked:", ranked)

my_cgpa = 8.9
rank = ranked.index(my_cgpa) + 1
print(f"CGPA {my_cgpa} → Rank {rank} out of {len(ranked)}")
```

```python
''' From a list of marks, use list comprehension to get: (a) all passing marks (>=40),
(b) all marks doubled, (c) only even marks.'''

marks = [78,89,56,36,89,34,80,33,66,22,55,21,11,60]

passing = [m for m in marks if m>=40]
print(passing)

doubled = [m*2 for m in marks]
print(marks)
print(doubled)

even = [m for m in marks if m%2==0]
print(even)
```

```python
# Create a list of student dictionaries. Find the topper and print all students sorted by marks descending.
student = [
    {"name":"Ronit","marks":97},
    {"name":"Aaryan","marks":94},
    {"name":"Rajan","marks":80},
    {"name":"Aaditya","marks":77},
    {"name":"Rahul","marks":99},
    {"name":"Sumit","marks":33},
    {"name":"Roshan","marks":53},
    {"name":"Keshav","marks":69},
]

ranked = sorted(student, key=lambda s: s["marks"], reverse=True)
print("Rank  Name      Marks")
print("-" * 25)
for i, stu in enumerate(ranked, start=1):
    print(f"  {i}   {stu['name']:<8}  {stu['marks']}")
 
print(f"Topper: {ranked[0]['name']} with {ranked[0]['marks']} marks") 

print(ranked)
```

```python
# Create a 3x3 marks matrix (3 students, 3 subjects). Print it as a table, then find the total for each student.
marks_matrix = [
    [85, 72, 91],   # Student 1: Aarav
    [60, 88, 75],   # Student 2: Priya
    [92, 65, 88],   # Student 3: Kabir
]
names = ["Aarav", "Priya", "Kabir"]
 
print(f"{'Name':<8} {'Math':>5} {'Physics':>8} {'CS':>4} {'Total':>7}")
print("-" * 38)
for i, row in enumerate(marks_matrix):
    total = sum(row)
    print(f"{names[i]:<8} {row[0]:>5} {row[1]:>8} {row[2]:>4} {total:>7}")
```
