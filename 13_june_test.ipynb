```python
def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Cannot divide"
    finally:
        print("Finally block")

result = divide(10, 0)
print(result)
```

```python
history = []

def add_turn(role, text):
    history.append({"role": role, "parts": [{"text": text}]})

add_turn("user", "Hello")
add_turn("model", "Hi there")
add_turn("user", "How are you?")
print(len(history))
print(history[1]["role"])
```

```python
def counter():
    n = 0
    def inc():
        nonlocal n
        n += 1
        return n
    return inc

c = counter()
print(c())
print(c())
print(c())
```

```python
def logger(func):
    def wrapper(*args):
        print("Calling:", func.__name__)
        return func(*args)
    return wrapper

@logger
def add(a, b):
    return a + b

print(add(3, 4))
```
