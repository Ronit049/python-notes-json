```python
def analyse(text):
    words = text.lower().split()
    sentences = [s.strip() for s in text.split(".") if len(s.strip()) > 5]
    return {"words": len(words), "sentences": len(sentences)}

result = analyse("Python is fun. I love coding.")
print(result["words"])
print(result["sentences"])
```

```python
def make_prompt(role, topic, audience, fmt):
    return f"""
    You are a {role}.
    Explain {topic} to a {audience}.
    Format: {fmt}
    """

p = make_prompt("tutor", "loops", "beginner", "code")
print("tutor" in p)
print("loops" in p)
print(type(p))
```
