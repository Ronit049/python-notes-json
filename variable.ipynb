```python
# List of subject names
subjects = ["Maths", "Physics", "Chemistry", "English", "Computer Science"]

# Print each subject with a serial number
for i in range(len(subjects)):
    print(f"{i + 1}. {subjects[i]}")
```

```python
marks = [33, 55, 43,87, 65, 56 ,42, 77, 99, 42]


passing_marks = [m for m in marks if m >= 40]
doubled_marks = [m * 2 for m in marks]
even_marks = [m for m in marks if m % 2 == 0]

print(passing_marks)
print(doubled_marks)
print(even_marks)
```

```python
# List of student dictionaries
students = [
    {'name': 'Ronit', 'marks': 85},
    {'name': 'Aman', 'marks': 92},
    {'name': 'Priya', 'marks': 78},
    {'name': 'Neha', 'marks': 95},
    {'name': 'Rahul', 'marks': 88}
]

sorted_students = sorted(
    students,
    key=lambda s: s['marks'],
    reverse=True
)

# Topper
topper = sorted_students[0]

print("Topper:")
print(topper)

print("\nStudents sorted by marks:")
for student in sorted_students:
    print(student)
```
