```python
def word_count(sentence):
    words = sentence.lower().split()
    freq = {}
    for word in words:
        freq[word] = freq.get(word,0)+1
    return freq

text = "the cat sat on mat the cat"
result = word_count(text)
print(result)
```

```python
def analyse_class(students):
    marks  = [s["marks"] for s in students]
    names  = [s["name"]  for s in students]
    avg    = round(sum(marks)/len(marks), 1)
    topper = students[marks.index(max(marks))]["name"]
    passed = len([m for m in marks if m >= 40])
    return {
        "average": avg,
        "topper" : topper,
        "pass_pct": round(passed/len(marks)*100, 1)
    }

data = [{"name":"Aarav","marks":88},
        {"name":"Priya","marks":95},
        {"name":"Kabir","marks":35}]
r = analyse_class(data)
print(f"Avg:{r['average']} Topper:{r['topper']}")
print(f"Pass rate: {r['pass_pct']}%")
```
